## **Stage 2 - Data Cleaning and Pre-processing**

Handle missing values (impute or drop)

Handle duplicates

```python
# Checking the Missing Values

print("Missing Values (Before):")
print(df.isnull().sum())
```

```python
# Handle Missing Values (Imputation)

# Numerical columns fill with mean

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Categorical columns fill with mode

cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# After Imputation checking the Missing Values

print("\nMissing Values (After):")
print(df.isnull().sum())
```

```python
# Handle Duplicates

print("\nDuplicate Rows (Before):", df.duplicated().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()

print("Duplicate Rows (After):", df.duplicated().sum())
```

```python
# Cleaned Data

display(df.head(50))
```

## **Stage 3 – EDA and Visualizations**

```python
import matplotlib.pyplot as plt

# 1. Screen Time vs Anxiety Score - Using Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df['Screen_Time_Hours'],
            df['Anxiety_Score'],
            color='blue',
            marker='o')

plt.title("Screen Time vs Anxiety Score")
plt.xlabel("Screen Time (Hours)")
plt.ylabel("Anxiety Score")
plt.grid(True)

plt.show()
```



*   Each dot represents one individual's screen time and corresponding anxiety score.

*   If the points show an upward trend (from left to right), it indicates a positive relationship, meaning individuals with higher screen time tend to have higher anxiety scores.

*   If the points show a downward trend, it indicates a negative relationship, meaning higher screen time is associated with lower anxiety scores.

*   If the points are scattered randomly without any clear pattern, it suggests little or no relationship between screen time and anxiety.

*   The spread of the points indicates the strength of the relationship. Closely clustered points suggest a stronger relationship and Widely scattered points suggest a weaker relationship.

```python
# 2. Wellbeing Index - Line Chart

plt.figure(figsize=(10,5))

plt.plot(df['Wellbeing_Index'],
         color='pink',
         marker='o',
         linewidth=2)

plt.title("Wellbeing Index of Participants")
plt.xlabel("Participant Index")
plt.ylabel("Wellbeing Index")

plt.grid(True)
plt.show()
```

*   The line chart shows that the Wellbeing Index varies across participants, indicating differences in their wellbeing levels.

*   Peaks represent participants with higher wellbeing scores, while dips indicate participants with lower wellbeing scores.

*   The chart does not show a consistent increasing or decreasing trend because the x-axis represents participant order, not time.

*   The line chart makes it easy to compare wellbeing scores across participants and identify those with relatively high or low wellbeing.


```python
# 3. Participants by Country - Bar Chart

country = df['Country'].value_counts()

plt.figure(figsize=(8,5))

plt.bar(country.index,
        country.values,
        color='orange')

plt.title("Participants by Country")
plt.xlabel("Country")
plt.ylabel("Count")

plt.show()
```

*   The bar chart shows the number of participants from each country included in the dataset.

*   The tallest bar represents the country with the highest number of participants, indicating the greatest representation in the study.

*   The shortest bar represents the country with the fewest participants, showing lower representation.

*   The chart allows for easy comparison of participant counts between countries, helping identify whether the sample is evenly distributed or concentrated in specific countries.

```python
# 4. Screen Time Distribution - Histogram

plt.figure(figsize=(8,5))

plt.hist(df['Screen_Time_Hours'],
         bins=8,
         color='skyblue',
         edgecolor='black')

plt.title("Distribution of Screen Time")
plt.xlabel("Screen Time Hours")
plt.ylabel("Frequency")

plt.show()
```

*   The histogram shows how participants are distributed across different ranges of daily screen time.

*   The tallest bar represents the screen time range with the highest number of participants, indicating the most common usage.

*   The distribution illustrates whether screen time values are concentrated within a few ranges or spread across many ranges.

*   If the histogram has a long tail or a few isolated bars, it may indicate participants with unusually high or low screen time compared to the rest of the group.

```python
# 5. Burnout Risk Distribution - Pie Chart

burnout = df['Burnout_Risk'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(burnout.values,
        labels=burnout.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Burnout Risk Distribution")

plt.show()
```

*   The pie chart shows the proportion of participants in each burnout risk category (e.g., Low, Medium, High).

*   The largest slice represents the burnout risk level with the highest percentage of participants, indicating the most common risk category.

*   The smallest slice represents the burnout risk level with the lowest percentage of participants, indicating the least common category.

*   The chart provides a clear visual comparison of burnout risk levels, helping to understand the overall distribution of burnout risk among participants.

```python
# 6. Sleep Quality Score - Box Chart

plt.figure(figsize=(6,5))

plt.boxplot(df['Sleep_Quality_Score'])

plt.title("Sleep Quality Score")
plt.ylabel("Score")

plt.show()
```

*   The box plot summarizes the distribution of Sleep Quality Scores, showing the central tendency and spread of the data.

*   The line inside the box represents the median sleep quality score, indicating the typical score among participants.

*   The height of the box (interquartile range) shows how much the middle 50% of sleep quality scores vary. A larger box indicates greater variability.

*   Any points plotted outside the whiskers represent outliers, indicating participants with unusually high or low sleep quality scores compared to the rest of the group.

```python
# 7. Content Preference - Horizontal Bar Chart

content = df['Content_Type_Preference'].value_counts()

plt.figure(figsize=(8,5))

plt.barh(content.index,
         content.values,
         color='purple')

plt.title("Content Type Preference")
plt.xlabel("Count")
plt.ylabel("Content Type")

plt.show()
```

*   The horizontal bar chart shows the number of participants preferring different types of content.

*   The longest bar represents the content type with the highest number of participants, indicating the most popular preference.

*   The shortest bar represents the content type with the lowest number of participants, showing the least common preference.

*   The chart allows easy comparison between different content categories and highlights variations in participants' content choices.

```python
# 8. Social Media Hours vs Scores - Scatter Plot

plt.figure(figsize=(8,5))

plt.scatter(df['Daily_Social_Media_Hours'],
            df['Anxiety_Score'],
            color='blue',
            label='Anxiety')

plt.scatter(df['Daily_Social_Media_Hours'],
            df['Wellbeing_Index'],
            color='Skyblue',
            label='Wellbeing')

plt.title("Social Media Hours vs Scores")
plt.xlabel("Daily Social Media Hours")
plt.ylabel("Score")

plt.legend()
plt.grid(True)

plt.show()
```

*   The scatter plot shows the relationship between daily social media usage hours and participants' anxiety scores and wellbeing index.

*   The distribution of anxiety score points helps identify whether increased social media usage is associated with higher or lower anxiety levels.

*   The wellbeing points show how social media usage may relate to participants' overall wellbeing, indicating possible positive or negative patterns.

*   The spread of points shows the strength of the relationship. A clear upward or downward pattern suggests correlation, while scattered points indicate a weaker relationship.

```python
# 9. Screen time across participants - Simple Area Chart

plt.figure(figsize=(10,5))

plt.fill_between(df.index,
                 df['Screen_Time_Hours'],
                 color='Green')

plt.title("Screen Time Hours")
plt.xlabel("Participant")
plt.ylabel("Hours")

plt.show()
```

*   The area chart compares the screen time hours of all participants, making it easy to observe differences in usage.

*   Higher peaks represent participants with more screen time, while lower peaks indicate participants with fewer hours of screen use.

*   The chart helps identify the overall pattern and variation in screen time, including participants with unusually high or low screen usage.

```python
# 10. Vertical stems with markers showing each participant's motivation level

plt.figure(figsize=(10,5))

markerline, stemlines, baseline = plt.stem(df['Motivation_Level'])

plt.setp(markerline, color='purple')
plt.setp(stemlines, color='pink')

plt.title("Motivation Level")
plt.xlabel("Participant")
plt.ylabel("Score")

plt.show()
```

*   The stem plot shows the motivation level score for each participant, allowing individual values to be viewed clearly.

*   Higher stems and markers represent participants with higher motivation levels, while shorter stems indicate lower motivation levels.

*   The visualization makes it easy to compare motivation scores across participants and observe the variation in motivation levels.

*   Any participants with exceptionally high or low scores can be easily identified, helping to detect trends or outliers in the dataset.

## **Executive Summary of EDA Findings**


This project explores the relationship between digital lifestyle habits and mental wellness among Gen Z individuals. Exploratory Data Analysis (EDA) was performed to identify patterns, trends, and relationships between factors such as screen time, social media usage, sleep quality, exercise, and stress levels.

**Key Findings :**

*   Individuals with higher daily screen time generally reported lower sleep
quality and higher stress levels.

*   Night-time social media usage was associated with reduced sleep quality and increased mental fatigue.

*   Participants who exercised more frequently tended to have better mental wellness scores and lower stress.

*   Content type preferences varied across users, showing different engagement behaviors among genders and countries.

*   Students and working professionals exhibited different digital usage patterns, with students spending comparatively more time on social media.

*   The dataset showed noticeable variations in digital habits across age groups, gender, and countries, highlighting diverse technology usage behaviors.

*   Correlation analysis indicated meaningful relationships between screen time, sleep quality, stress level, and overall wellness indicators.

**Overall Summary :**

The EDA suggests that balanced digital habits, sufficient sleep, and regular physical activity are positively associated with better mental well-being among Gen Z. Excessive screen time and frequent late-night social media use appear to negatively impact sleep quality and stress levels. These findings provide valuable insights into the interaction between technology use and mental health, supporting data-driven awareness and healthier digital lifestyle practices.

## **Key Business Insights :**


*   Higher screen time is associated with lower sleep quality, indicating that excessive device usage can negatively affect overall well-being.

*   Frequent night-time social media use is linked to higher stress levels, emphasizing the importance of healthy digital habits.

*   Individuals who exercise regularly generally report better mental wellness scores, showing the positive impact of physical activity on mental health.

*   Digital lifestyle patterns vary across age, gender, student/working status, and country, enabling targeted wellness strategies for different user groups.

*   The findings can help organizations, educational institutions, and digital platforms design personalized wellness programs and responsible technology usage features to promote healthier digital lifestyles.

## **Strategic Recommendations**
